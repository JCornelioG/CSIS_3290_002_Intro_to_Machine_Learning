### Introduction to Matplotlib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. 
After its inception, Matplotlib quickly grew in popularity within the scientific Python community. Developers and researchers began contributing to its development, enhancing its capabilities, and extending its functionality. 

Matplotlib's ability to create high-quality, publication-ready graphics has had a significant impact on Python's popularity in scientific computing and data science. It has enabled Python to serve as a viable free and open-source alternative to commercial software like MATLAB. 


````python
import numpy as np
import pandas as pd
# Importing Matplotlib
import matplotlib
import matplotlib.pyplot as plt

# Check Matplotlib version
print("Matplotlib version:", matplotlib.__version__)

# this magic function is necessary in the latest jupyter notebook
%matplotlib inline  
````

## Template of using `plt`

Using `plt` is generally following the structure:
    
    plt.figure()
    plt.plot(____________)
    plt.show()
    
1. **Setting Up the Figure**:
The `plt.figure()` function is your starting point in Matplotlib plotting. It initializes a new figure object—a container that will hold all elements of your plot such as axes, graphics, text, and labels. You can think of it as a blank canvas upon which your data will be plotted. Optionally, you can specify attributes of the figure, such as its size and resolution, using parameters like figsize and dpi.
```python
plt.figure(figsize=(10, 6))  # Creates a figure 10 inches wide by 6 inches tall
```

2. **Creating the Plot**:
The `plt.plot()` function is versatile, allowing for simple line plots as well as customization to fit specific needs. At its simplest, plt.plot() requires data points that you want to plot. You can provide these points as a list or a NumPy array; typically, these points represent the y-values to be plotted against their corresponding x-values. The default of `plt.plot` is to draw a line plot.
```python
# Plotting a simple line plot
x = [1, 2, 3, 4]  # Define x-axis points
y = [1, 4, 9, 16]  # Define y-axis points
plt.plot(x, y)
```

3. **Displaying the Plot**:
The `plt.show()` function displays the plot. In a Jupyter Notebook, this function isn’t always necessary because plots can be set to display automatically. However, using `plt.show()` is a good practice as it makes your code transferrable to other Python environments and ensures that the plot is rendered at this point in your script.
```python
plt.show()
```

## Basic Plotting

Learn the basics of creating and customizing plots with Matplotlib. This section covers simple line plots and how to adjust key plot elements like labels, titles, and legends.

````python
# a simple example
plt.figure()
plt.plot([1, 2, 4, 6, 10]) # only y-coordinate, the default x-coordinate will be 0, 1, 2, 3,...
plt.show()
````

````python
plt.figure()
plt.plot([2, 4, 6, 8, 10], [1, 2, 5, 6, 10]) # passing in both x and y coordinates
plt.show()
````

````python
# Basic line plot
plt.figure(figsize=(8, 6))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Line 1')
plt.title('Basic Line Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend()
plt.show()
````

### Explanation of the previous code example

**Step 1: Creating a Figure.** Every plot in Matplotlib resides within a Figure object. Here, we start by creating a new figure with a specific size. The figsize parameter is a tuple that defines the width and height of the figure in inches.
```python
plt.figure(figsize=(8, 6))
```
**Step 2: Plotting the Data.**. We plot a simple line graph using the plot function. The first list `[1, 2, 3, 4]` represents the x-values, while the second list `[1, 4, 9, 16]` represents the y-values. We also add a label for the line, which will be used in the legend.
```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Line 1')
```
**Step 3: Adding a Title and Labels.**
It's important to label your graph so that others can understand what it represents. title adds a title to the plot, and xlabel and ylabel label the x-axis and y-axis, respectively.
```python
plt.title('Basic Line Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
```
**Step 4: Adding a Legend.**
A legend is crucial when your plot has more than one line or element. It helps distinguish between the different data series on the graph. Here, we use the legend function to display the legend, which automatically uses the labels specified in the plot function.
```python
plt.legend()
```
**Step 5: Displaying the Plot.**
Finally, use the show function to display the plot. In a Jupyter Notebook, this will render the plot right under the cell where the code is executed.
```python
plt.show()
```

## Advanced Plot Customization

Matplotlib charts have more customization options, including setting axes scales, modifying annotation, and applying themes for more professional presentations of data.


````python
# Advanced plot customization
plt.figure(figsize=(8, 6))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r--', label='Red Dashed Line')
plt.yscale('log')
plt.title('Advanced Customization')
plt.xlabel('X-axis')
plt.ylabel('Logarithmic Y-axis')
plt.grid(True)
plt.legend()
plt.show()
````

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. This section covers some of the fundamental plotting functions available in Matplotlib: `plt.scatter`, `plt.hist`, `plt.bar`, and `plt.boxplot`. Each type of plot can be used to highlight different aspects of your data. 


### Scatter Plots

Scatter plots are ideal for visualizing the relationships between two numerical variables. They show how much one variable is affected by another, which helps in identifying correlations and trends.


````python
# Generating sample data
x = np.random.rand(50)
y = np.random.rand(50)

# Creating a scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('Random x values')
plt.ylabel('Random y values')
plt.show()
````

### Histograms

Histograms are great for illustrating the distribution of data points. They provide a sense of the density of values, showing the frequency of data points within certain ranges.


````python
# Generating sample data
data = np.random.randn(1000)

# Creating a histogram
plt.hist(data, bins=30)
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
````

### Bar Charts

Bar charts are useful for comparing quantities across different categories. They are one of the most common types of graphical representation used in data analysis.


````python
# Categories and values
categories = ['Cat', 'Dog', 'Bird']
values = [15, 18, 6]

# Creating a bar chart
plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

### Box Plots

Box plots, also known as box-and-whisker plots, summarize data through quartiles. They highlight the median, the upper and lower quartiles, and potential outliers. This makes them extremely useful for statistical analyses of data distributions.


````python
# Generating sample data
data = np.random.rand(50)

# Creating a box plot
plt.boxplot(data)
plt.title('Box Plot Example')
plt.ylabel('Random values')
plt.show()
````

## Saving Figures

Learn how to save your Matplotlib figures in various formats for use in reports, presentations, or web applications.


````python
# Saving a figure
plt.figure(figsize=(8, 6))
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Figure to Save')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('saved_figure.png')  # Save the figure
plt.show()
````

## Introduction to Plotting with Pandas

Pandas provides built-in plotting capabilities that are built on `Matplotlib`, allowing for quick and straightforward data visualization directly from DataFrames and Series. This section introduces the basics of plotting with pandas, including setting up your environment and a quick overview of the types of plots you can create.


### Basic Plotting with Pandas

Learn how to create simple line plots using pandas. This section covers how to plot data directly from DataFrames and customize basic plot elements.


````python
# Create a simple DataFrame
df = pd.DataFrame({
    'x': range(10),
    'y': range(10),
    'y2': [x**2 for x in range(10)]
})

# Simple line plot
df.plot(x='x', y='y', label='Linear', title='Basic Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
````

### Plotting Multiple Columns

Pandas allows for easy plotting of multiple columns. This section demonstrates how to plot multiple sets of data on the same chart and how to customize the legends and colors.


````python
# Plotting multiple columns
ax = df.plot(x='x', y='y', label='Linear', figsize=(10, 5))
df.plot(x='x', y='y2', label='Quadratic', ax=ax)
ax.set_title('Multiple Columns Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
plt.show()
````

### Creating Bar Plots

Bar plots are useful for comparing quantities corresponding to different groups. This section covers how to create vertical and horizontal bar plots from DataFrame data.


````python
# Creating a bar plot
df.plot(kind='bar', x='x', y='y2', title='Bar Plot Example')
plt.xlabel('X axis')
plt.ylabel('Y axis squared')
plt.show()
````

### Advanced Pandas Plotting Features

Explore more advanced plotting features in pandas, such as histograms, box plots, and area plots. This section also introduces how to customize plots with additional Matplotlib parameters.


````python
# Histogram
df['y2'].plot(kind='hist', title='Histogram of y2')
plt.xlabel('Y axis squared values')
plt.show()

# Box plot
df[['y', 'y2']].plot(kind='box', title='Box Plot Example')
plt.show()

# Area plot
df.plot(kind='area', stacked=False, title='Area Plot Example')
plt.ylabel('Values')
plt.show()
````

## Subplots

Matplotlib's `subplots` function is a powerful tool that allows you to organize multiple plots in a single figure. Each subplot is placed in a grid within the figure, and you can control the number of rows and columns as well as the spacing between the plots. This section introduces the basic usage of `subplots`.


````python
import matplotlib.pyplot as plt

# Creating a single subplot
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax.set_title('Simple Plot')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
plt.show()
````

### Creating Multiple Subplots

We can create a figure with multiple subplots arranged in a grid. This is useful for comparing different datasets side-by-side or showing different aspects of the same data.


````python
# Creating a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))  # 2 rows, 2 columns
axs[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axs[0, 0].set_title('Plot 1')
axs[0, 1].plot([1, 2, 3, 4], [1, 4, 6, 8])
axs[0, 1].set_title('Plot 2')
axs[1, 0].plot([1, 2, 3, 4], [1, 2, 3, 4])
axs[1, 0].set_title('Plot 3')
axs[1, 1].plot([1, 2, 3, 4], [10, 20, 30, 40])
axs[1, 1].set_title('Plot 4')
plt.tight_layout()  # Adjust subplots to fit into figure area.
plt.show()
````

### Customizing Subplots

We can customize the layout and appearance of subplots to improve readability and presentation. Learn how to adjust spacing, align axes, and apply consistent styles across all subplots.


````python
# Adjusting spacing and aligning axes
fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex='col', sharey='row')
fig.subplots_adjust(hspace=0.5, wspace=0.5)  # Adjust horizontal and vertical spacing
axs[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axs[0, 1].plot([1, 2, 3, 4], [1, 4, 6, 8])
axs[1, 0].plot([1, 2, 3, 4], [1, 2, 3, 4])
axs[1, 1].plot([1, 2, 3, 4], [10, 20, 30, 40])
plt.show()
````

````python
# Example of a complex grid and combining plots
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(221)  # Top left corner
ax2 = fig.add_subplot(222)  # Top right corner
ax3 = fig.add_subplot(212)  # Bottom span

ax1.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax1.set_title('Top Left Plot')
ax2.plot([1, 2, 3, 4], [1, 4, 6, 8])
ax2.set_title('Top Right Plot')
ax3.plot([1, 2, 3, 4], [10, 20, 30, 40])
ax3.set_title('Bottom Span Plot')
plt.tight_layout()
plt.show()
````

### Subplots in Pandas

````python
# Create a sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

# Create subplots from the DataFrame
axs = df.plot(subplots=True, layout=(3, 1), figsize=(10, 10), sharex=True)
````

You can further customize each subplot by interacting with each axis object returned by the .plot() method.

````python
# Customize each subplot
axs = df.plot(subplots=True, layout=(3, 1), figsize=(5, 5), sharex=True, title='Multiple Plots')
for ax in axs:
    ax[0].set_ylabel('Value')
    ax[0].legend(loc='upper right')
plt.tight_layout()  # Adjust subplot parameters to give sufficient space
````
